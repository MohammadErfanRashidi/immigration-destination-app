```python
# Create initial dataframe
all_countries = sorted(set().union(*[set(lst) for lst in data.values()]))

df = pd.DataFrame({'country': all_countries})

df.head()
```

```python
# Check if we have duplicates
df.shape[0] == df['country'].nunique()
```

```python
# Create English status
def engish_status(country):
    if country in set(data['English speaking']):
        return 'first'
    elif country in set(data['English as second language']):
        return 'second'
    else:
        return 'not dominant'

df['english_status'] = df['country'].apply(engish_status)

df['english_status'].value_counts()
```

```python
df.head()
```

```python
# Safest countries
df['safest'] = df['country'].apply(lambda x: 1 if x in set(data['Safest countries 2025']) else 0)

df['safest'].value_counts()
```

```python
# Least racist
df['least_racist'] = df['country'].apply(lambda x: 1 if x in set(data['Least racist countries 2026']) else 0)

df['least_racist'].value_counts()
```

```python
# Best health care
df['best_healthcare'] = df['country'].apply(lambda x: 1 if x in set(data['Best health care']) else 0)

# Wage vs. living cost
df['best_wage_cost_balance'] = df['country'].apply(lambda x: 1 if x in set(data['Wage vs living cost']) else 0)

# Most Iranians
df['most_iranians'] = df['country'].apply(lambda x: 1 if x in set(data['Countries with largest Iranians']) else 0)

# High paying psychology
df['highest_paying_psychology'] = df['country'].apply(lambda x: 1 if x in set(data['Highest paying psychologists']) else 0)

# Top education
df['top_education'] = df['country'].apply(lambda x: 1 if x in set(data['Top education']) else 0)

# Best immigration policies
df['best_immigration_policies'] = df['country'].apply(lambda x: 1 if x in set(data['Best immigrant policies']) else 0)
```

```python
df.head()
```

```python
# Passport 
top_passports_explicit = set(data['Top passports']) - {'EU'}

top_passport_set = top_passports_explicit.union(eu_countries)

df['top_passport'] = df['country'].apply(lambda x: 1 if x in top_passport_set else 0)

df['top_passport'].value_counts()
```

```python
# Highest job opportunities
df['highest_job_opportunities'] = df['country'].apply(lambda x: 1 if x in set(data['Most job opportunities']) else 0)

# Most migrant acceptance
df['most_migrant_acceptance'] = df['country'].apply(lambda x: 1 if x in set(data['Migrant acceptance']) else 0)

# Best healthcare cost comparing to salary
df['best_healthcare_cost_to_salary'] = df['country'].apply(lambda x: 1 if x in set(data['Best health care cost vs salary']) else 0)

# Lowest tolerance on drugs
df['lowest_tolerance_on_drugs'] = df['country'].apply(lambda x: 1 if x in set(data['Lowest tolerance drugs']) else 0)

# Most annual reading
df['highest_annual_reading_rate'] = df['country'].apply(lambda x: 1 if x in set(data['Most annual reading']) else 0)

# Best mental healthcare
df['best_mental_health_care'] = df['country'].apply(lambda x: 1 if x in set(data['Best mental health care']) else 0)

# Most psychologists per capita
df['most_psychologists_per_capita'] = df['country'].apply(lambda x: 1 if x in set(data['Most psychologists per capita']) else 0)

# Healthiest 
df['healthiest'] = df['country'].apply(lambda x: 1 if x in set(data['Healthiest']) else 0)

# Most prosperous 
df['most_prosperous'] = df['country'].apply(lambda x: 1 if x in set(data['Most prosperous']) else 0)

# Best public transport
df['best_public_transport'] = df['country'].apply(lambda x: 1 if x in set(data['Best public transport']) else 0)

# Cleanest
df['cleanest'] = df['country'].apply(lambda x: 1 if x in set(data['Cleanest']) else 0)

# Pregnancy leave paid above 20 weeks
df['pregnancy_leave_above_20_weeks'] = df['country'].apply(lambda x: 1 if x in set(data['Pregnancy leave paid above 20 weeks']) else 0)

# Best work life balance
df['best_work_life_balance'] = df['country'].apply(lambda x: 1 if x in set(data['Best work life balance']) else 0)

# Tax above 40 percent
df['tax_above_40_percent'] = df['country'].apply(lambda x: 1 if x in set(data['Tax above 40 percent']) else 0)

df.head()
```

```python
# Add cat columns
df['continent'] = df['country'].map(continent_map)
df['climate'] = df['country'].map(climate_map)
df['religion'] = df['country'].map(religion_map)
df['race'] = df['country'].map(race_map)

# New head
df.head()
```
